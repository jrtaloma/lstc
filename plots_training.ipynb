```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## NonInvasiveFetalECGThorax

### RI

```python
df_ri = pd.read_csv('./csv_logs/NonInvasiveFetalECGThorax1_RI.csv')
display(df_ri)
```

```python
df_ri = df_ri[df_ri.columns[[0,4,10,16,22,28,34]]]
column_names = list(df_ri.columns.values)
print(column_names)
```

```python
df_ri.rename(columns={column_names[1]: 'LoSTer_F', column_names[2]: 'IDEC', column_names[3]: 'DTCC', column_names[4]: 'LoSTer', column_names[5]: 'CKM', column_names[6]: 'DTCR'}, inplace=True)
df_ri.set_index('Epoch', inplace=True)
df_ri = df_ri.loc[:, ['LoSTer', 'LoSTer_F', 'CKM', 'IDEC', 'DTCR', 'DTCC']]
display(df_ri)
```

```python
df_ri.plot()
```

### NMI

```python
df_nmi = pd.read_csv('./csv_logs/NonInvasiveFetalECGThorax1_NMI.csv')
display(df_nmi)
```

```python
df_nmi = df_nmi[df_nmi.columns[[0,4,10,16,22,28,34]]]
column_names = list(df_nmi.columns.values)
print(column_names)
```

```python
df_nmi.rename(columns={column_names[1]: 'LoSTer_F', column_names[2]: 'IDEC', column_names[3]: 'DTCC', column_names[4]: 'LoSTer', column_names[5]: 'CKM', column_names[6]: 'DTCR'}, inplace=True)
df_nmi.set_index('Epoch', inplace=True)
df_nmi = df_nmi.loc[:, ['LoSTer', 'LoSTer_F', 'CKM', 'IDEC', 'DTCR', 'DTCC']]
display(df_nmi)
```

```python
df_nmi.plot()
```

### Plots

```python
model_names = ['LoSTer', 'LoSTer (soft $\it{k}$-means)', 'CKM', 'IDEC', 'DTCR', 'DTCC']
```

```python
def plot_curves(ax, df, show_legend=False):
    for i in range(len(model_names)):
        ax.plot(np.arange(0, len(df)), df.values[:,i])
    if show_legend:
        ax.legend(model_names, fontsize=14)
```

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 5), dpi=300, sharex=False)

# Set horizontal and vertical spacing
fig.subplots_adjust(wspace=0.2, hspace=0)

# Metrics
metrics = ['RI', 'NMI']

# Plot curves
plot_curves(axs[0], df_ri, show_legend=False)
plot_curves(axs[1], df_nmi, show_legend=False)

# Adjusting the axes and setting ticks
for i in range(axs.shape[0]):
    axs[i].set_ylabel(metrics[i], fontsize=16)
    axs[i].set_xlabel('Epoch', fontsize=16)
    axs[i].tick_params(axis='both', which='major', labelsize=14)

# Create legend handles
handles = [plt.Line2D([0], [0], color=f'C{m}', lw=4) for m in range(len(model_names))]

# Add a single shared legend to the figure
fig.legend(handles, model_names, loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=len(model_names), frameon=False, fontsize=16)

# Save figure
plt.savefig('./figures/plots_NonInvasiveFetalECGThorax1.pdf', format='pdf', bbox_inches='tight')

plt.show()
```
